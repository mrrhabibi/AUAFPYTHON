```python
#This is how we get info from the user and check if it's true or false according to out conditons
#By using input() we can get info from the user
username = input('USERNAME')
password = input('PASSWORD')
#These are our conditions
# (:) this mark should be used in the end of each condition
if username==('Mansoor') and password==('Mansoor123'):
    print('You are successfully logged in.')
else:
    print('You are not able to log in')
```

```python
#this is how we can use range of numbers instead of sets...
#range(start,end,stepover)
total_sum = 0
for items in range(1,20):
    total_sum = items + total_sum
    
print(total_sum)    
```
